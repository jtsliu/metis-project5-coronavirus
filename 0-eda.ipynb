```python
import pandas as pd
import geopy
import geopandas
import folium
from folium.plugins import FastMarkerCluster
```

```python
confirmed_df = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv')
deaths_df = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv')
recovered_df = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv')
```

```python
current_df = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_daily_reports/03-12-2020.csv')
```

```python
world_map = folium.Map(
    min_zoom=1,
    tiles='cartodbpositron'
)
current_df.apply(lambda row:folium.CircleMarker(location=[row["Latitude"], row["Longitude"]],
                                                popup=f"{row['Province/State']} \n \
                                                Country: {row['Country/Region']}\n \
                                                Confirmed: {row['Confirmed']}   \n \
                                                Deaths: {row['Deaths']}         \n \
                                                Recovered: {row['Recovered']}",
                                                radius=(row['Confirmed']/current_df.Confirmed.sum())*50,
                                                color='crimson',
                                                fill=True,
                                                fill_color='crimson'
                                               ).add_to(world_map), axis=1)

# for i in range(0,len(current_df)):
#     folium.CircleMarker(
#       location=[current_df.iloc[i]['Latitude'], current_df.iloc[i]['Longitude']],
#       popup=current_df.iloc[i]['Country/Region'],
#       radius=current_df.iloc[i]['Confirmed']*10000,
#       color='crimson',
#       fill=True,
#       fill_color='crimson'
#     ).add_to(world_map)


# FastMarkerCluster(data=list(zip(current_df['Latitude'].values, current_df['Longitude'].values))).add_to(world_map)
# folium.LayerControl().add_to(world_map)

world_map
```
