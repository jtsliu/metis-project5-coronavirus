```python
import pandas as pd
import pickle
```

```python
# Source: https://github.com/grammakov/USA-cities-and-states
```

```python
df = pd.read_csv('tweets/us_cities_states_counties.txt', delimiter='|')
```

```python
df
```

```python
df['State full'] = df['State full'].apply(str)
df['State full'] = df['State full'].apply(lambda x: x.lower())
df['State full'] = df['State full'].apply(lambda x: x.replace(' ', ''))


df['City'] = df['City'].apply(str)
df['City'] = df['City'].apply(lambda x: x.lower())
```

```python
df
```

```python
state_list = ['california', 'colorado', 'florida',
              'georgia', 'idaho', 'illinois',
              'louisiana', 'massachusetts', 'newyork',
              'tennessee', 'texas', 'washington']

df = df.loc[df['State full'].isin(state_list)]
df.reset_index(drop=True, inplace=True)
```

```python
# df.to_csv('tweets/selected_us_cities_states_counties.txt', index=False)
```

```python
cities = df.groupby(['State full'])['City'].apply(set).to_dict()
```

```python
pickle.dump(cities, open('tweets/cities_list.txt', 'wb'))
# cities = pickle.load(open('tweets/cities_list.txt', 'rb'))
```
